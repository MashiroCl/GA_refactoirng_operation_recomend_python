```python
import ast
file = "/Users/leichen/Desktop/GA.txt"
    
with open(file) as f:
    lines = f.readlines()
data = [ast.literal_eval(line) for line in lines]

qmood = {}
qmood["Effectiveness"]=[]
qmood["Extendibility"]=[]
qmood["Flexibility"]=[]
qmood["Functionality"]=[]
qmood["Resusability"]=[]
qmood["Understandability"]=[]
qmood["HighestOwnership"]=[]
qmood["NumOfCommiter"]=[]
for each in data:
    qmood["Effectiveness"].append(each[0]) 
    qmood["Extendibility"].append(each[1])
    qmood["Flexibility"].append(each[2])
    qmood["Functionality"].append(each[3])
    qmood["Resusability"].append(each[4])
    qmood["Understandability"].append(each[5])
    qmood["HighestOwnership"].append(each[6])
    qmood["NumOfCommiter"].append(each[7])
    
```

```python
def readData(file):
    with open(file) as f:
        lines = f.readlines()
    data = []
    for each in lines:
        data.append(each.split(" "))
    qmood = {}
    qmood["Effectiveness"]=[]
    qmood["Extendibility"]=[]
    qmood["Flexibility"]=[]
    qmood["Functionality"]=[]
    qmood["Resusability"]=[]
    qmood["Understandability"]=[]
    qmood["HighestOwnership"]=[]
    qmood["NumOfCommiter"]=[]
    for each in data:
            qmood["Effectiveness"].append(each[0]) 
            qmood["Extendibility"].append(each[1])
            qmood["Flexibility"].append(each[2])
            qmood["Functionality"].append(each[3])
            qmood["Resusability"].append(each[4])
            qmood["Understandability"].append(each[5])
            qmood["HighestOwnership"].append(each[6])
            qmood["NumOfCommiter"].append(each[7])
    return qmood
```

```python
import matplotlib.pyplot as plt
def draw_pic(qmood,name="result"):
    count = 1
#     x=range(len(qmood["Effectiveness"])
#     plt.subplot(2,4,count).title.set_text("Effectiviness")
#     plt.plot(x,qmood["Effectiveness"])
    plt.tight_layout()
    for each in qmood:
        x=range(len(qmood[each]))
        plt.subplot(2,4,count).title.set_text(each)
        plt.plot(x,qmood[each])
        count=count+1
    plt.subplots_adjust(left=0,right=1,wspace =1, hspace =1)
    plt.savefig( "/Users/leichen/Desktop/"+name+".png",dpi=500, bbox_inches = "tight")
```

```python
file = "/Users/leichen/Desktop/FUN.NSGAII.SubsetSum"
qmood = readData(file)
draw_pic(qmood,"20211106")
# for each in qmood:
#     print(len(qmood[each]))
```
